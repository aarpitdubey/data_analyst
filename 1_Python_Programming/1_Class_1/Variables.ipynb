# Python Variables

#### Data/value can be stored in "Temporary Storage Spaces" called Variables.

Question - What is data?

Example -  What is your fav. color?
Color choices - Blue, Green, Yellow these are some data

- Like when someone ask you what's your favorite color you said it's Blue color and then your memories stores that color then,
- After a month someone ask again the same question but now, your mood changes and then you said it's Green color and again your memories replace previous color and store this one.
- Then, again after a month some ask the same question and now, your favorite color becomes Yellow then again in your memory it's replace previous color and stores Yellow color.

Question - In programming, how you can store and remember?

- To store the favorite color in human we have brain but in programming we have a concept of variables.
- Variable are use to store data and also, we can used the store data through variables

```python

favorite_color = "Blue"

```

````python
favorite_color = "Blue"

print(favorite_color)

after_1_month_favorite_color = "Green"

print(after_1_month_favorite_color)

after_2nd_month_favorite_color = "Yellow"

print(after_2nd_month_favorite_color)
````

On the other hand, you can said Variable as empty bucket (or) as a empty box which can store the data.

- Variable provides a storage space for data.
- We can create multiple variable for various type of data (or) we can use same variable for various data but remember whenever we used same variable then every time we replace the previous data with new data.

````python
# favorite_color is a empty variable and we filled the memory space with "Blue" 
favorite_color = "Blue"
````

````python
# This statement shows how our variable remember the data
favorite_color
````

````python
# Now, we have to change the data with "Green"
favorite_color = "Green"
````

- Now, the new data replace the stored data and variable name remains the same

- let's check what will be it's output?

````python
favorite_color
````

- As we can see the previous data "Blue" replaced by new data "Green"

- Variable is a temporary store which provide storage space for data.

# Data Types in Python

#### Each variable is associated with a data type



````python
student_name = "Ravi" # String (or) str data type
roll_number  = 12 # Integer (or) int data type
percentage   = 93.5 # float data type
location     = "Bangalore" # String (or) str data type

print(f"\nHi there, \n\nI am {student_name}, my roll number is {roll_number}\nI have got {percentage}% in 12th standard and\ncurrently persuing my graduation from {location} university.")
````

````python
is_sunny   = False
is_raining = True

print(f"\nWhat's the weather of today?\n\nIs it sunny? {is_sunny}\n\n or \n\nis it raining? {is_raining}")
````

 - Remember, we don't have to explicitly mentioned about any of the data type.
 
 - Like in c, c++ or java we have mentioned about the data type before variable.
 
### In C language Variable declaration: 
```c
#include <stdio.h>

int main() {
    int age = 12;  // Variable declaration
    float height = 73.5;  // Variable declaration
    
    // Rest of the program
    return 0;
}
```
### In C++ language Variable declaration: 
```c++
#include <iostream>
using namespace std;

int main() {
    int age = 12;  // Variable declaration
    double height = 73.5;  // Variable declaration
    
    // Rest of the program
    return 0;
}

```
### In Java language Variable declaration: 
```java
public class Main {
    public static void main(String[] args) {
        int age = 12;  // Variable declaration
        double height = 73.5;  // Variable declaration
        
        // Rest of the program
    }
}

```

# Python Tokens

### Tokens are the smallest meaningful component in a Program



````python
number1 = 10
number2 = 5

result = number1 + number2

print("The result is:", result)
````

In this code, the tokens include:

Keywords: `print`

Identifiers: `number1`, `number2`, `result`

Operators: `=`, `+`

Literals: `()`, `:`, `"`
