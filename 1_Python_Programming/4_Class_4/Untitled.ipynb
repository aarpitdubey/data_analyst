# Python OOPs (Object Oriented Programming)

- OOps is like building a Car using modular parts. Here, code is broken down into smaller, reusable pieces called objects that represent real-world entities.

## Classes

- Class is a template/blue-print for real-world entities






```python
class MobilePhone:
    
    def __init__(self, model, color, cost):
        self.model=model
        self.color=color
        self.cost=cost
        
    def play_game(self):
        return f'{self.model}, has a color {self.color} and have a cost {self.cost} '
    
    def make_call(self):
        return f'Make call'
```

```python
mp = MobilePhone("Toyota", "Black", "14 Lac")
mp.play_game()
```

```python
# Creating the 'Phone' class
class Phone:
    
    def make_call(self):
        print("Making Phone call")
        
    def play_game(self):
        print("Playing Game")
```

```python
p1 = Phone() # instantiating the 'p1' object
```

```python
p1.make_call() # invoking methods through object 'p1'
```

```python
p1.play_game() # invoking methods through object 'p1'
```

```python
p2 = Phone()
```

```python
p2.make_call()
```

```python
p2.play_game()
```

```python
class Phone:
    
    def set_color(self, color):
        self.color=color
        
    def show_color(self):
        return self.color
    
    def make_call(self):
        print("I am making a call")
        
    def play_game(self):
        print("I am playing a game")
```

```python
p1 = Phone()
```

```python
p1.set_color("red")
```

```python
p1.show_color()
```

```python
p2 = Phone()
```

```python
p2.set_color("green")
```

```python
p2.show_color()
```

```python
class Student:
    
    def __init__(self, name, age, student_id):
        self.name       = name
        self.age        = age
        self.student_id = student_id
        
    def display_info(self):
        print("Name of the student is", self.name)
        print("Age of the student is", self.age)
        print("ID of the student is", self.student_id)
```

```python
s1 = Student("Alice", 20, "1234")
```

```python
s2 = Student("Bob", 25, "1111")
```

```python
s1.display_info()
```

```python
s2.display_info()
```

## Objects

- Objects are specific instances of a class





```python
class Phone:
    
    def __init__(self, model, color, cost):
        self.model = model
        self.color = color
        self.cost  = cost
        
    def get_color(self):
        return self.color
    
    def make_call(self):
        return f"You are making call through {self.model} phone."
    
    def playing_game(self):
        return f"You are playing game on {self.model} phone"
    
    def phone_price(self):
        return self.cost
    
```

```python
apple = Phone("Apple", "Silver", "1,45000")
```

```python
apple.model
```

```python
apple.get_color()
```

```python
apple.make_call()
```

```python
apple.phone_price()
```

```python
motorola = Phone("Motorola", "Black", "33000")
```

```python
motorola.make_call()
```

```python
motorola.phone_price()
```

```python
motorola.get_color()
```

```python
samsung = Phone("Samsung", "White", "56000")
```

```python
samsung.get_color()
```

```python
samsung.make_call()
```

```python
samsung.playing_game()
```

**Note :** These apple, motorola, samsung are the object of the class `Phone` and you can see how we can use the Attributes and method of a class using these objects or instances.
